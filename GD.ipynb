```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
#Lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red')
```

```python
#lets assume m=78.35
#and for first iteration b=0
y_pre = ((78.35*X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pre,color='#00a65a',label='b = 0')
```

```python
m = 78.35
b = 0
slope = -2 * np.sum(y-m*X.ravel()-b)
slope
```

```python
a = .1
step_size = a*slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pre1 = ((78.35*X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pre1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pre,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
b = -100
m=78.35
a=.01

epochs = 10

for i in range(epochs):
  slope = -2*np.sum(y-m*X.ravel() - b)
  b = b - (a*slope)

  y_pre = m*X + b
  plt.plot(X,y_pre)

plt.scatter(X,y)
```

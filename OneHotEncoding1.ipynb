```python
import pandas as pd
```

```python
df = pd.read_csv('/content/homeprices.csv')
df
```

```python
#Using pandas to get dummy variablesm
```

```python
dummy = pd.get_dummies(df.town)
dummy
```

```python
merged = pd.concat([df,dummy],axis='columns')
merged
```

```python
final = merged.drop(['town'],axis='columns')
final
```

```python
#dummy variable trap
```

```python
finall = final.drop(['west windsor'],axis='columns')
finall
```

```python
X = finall.drop(['price'],axis='columns')
X
```

```python
y = finall.price
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
model.predict([[3400,0,0]]) #3400 sqr ft in west windsor
```

```python
#using sklearn OneHotEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
X = dfle[['town','area']].values
X
```

```python
y = dfle.price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town',OneHotEncoder(),[0])],remainder='passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]]) #3400 sqr ft home in west windsor
```
